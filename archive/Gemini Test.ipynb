Gemini Call

```python
import requests
import json
# Gen AI packages Gemini
from google import genai
from google.genai import types
```

```python
# The client gets the API key from the environment variable `GEMINI_API_KEY`.
token = ''
client = genai.Client(api_key=token)

def chart_analysis(prompt, image):
    response = client.models.generate_content(
        model="gemini-2.5-flash",
        contents=[prompt],
        config=types.GenerateContentConfig(
            system_instruction='You are a helpful assistant',
            temperature=0.1
        )
    )
    return response.text
```

Oanda

```python
import v20
from datetime import datetime
import time
from dateutil.relativedelta import relativedelta
```

```python
# Practice Acct No:
# Practice API Token:
acct = ''
token = ''

# REST API - historical price action, place orders, account info, etc
rest_url = 'api-fxpractice.oanda.com'
```

```python
# Function definition to convert json to pandas dataframe
def json_to_pandas(json):
    json_file = json.json()
    price_json = json_file['candles']
    times = []
    close_price, high_price, low_price, open_price = [], [], [], []

    for candle in price_json:
        times.append(candle['time'])
        close_price.append(float(candle['mid']['c']))
        high_price.append(float(candle['mid']['h']))
        low_price.append(float(candle['mid']['l']))
        open_price.append(float(candle['mid']['o']))

    df = pd.DataFrame({'datetime': pd.to_datetime(times).tz_localize(None),
                       'close': close_price,
                       'high': high_price,
                       'low': low_price,
                       'open': open_price})
    df.index = pd.to_datetime(times).tz_localize(None)
    return df
```

```python
# for currency in li_currency:
currency = 'GBP_USD'
interval = {'M15'}
print(f'---------- {currency} ----------')
# Get current prices
rest = v20.Context(hostname=rest_url, token=token, port=443)

response_rest = rest.pricing.get(accountID=acct, instruments=currency)
json.loads(response_rest.body["prices"][0].json())
bid_at_trade = json.loads(response_rest.body["prices"][0].json())['bids'][0]['price']
ask_at_trade = json.loads(response_rest.body["prices"][0].json())['asks'][0]['price']
spread_at_trade = round(abs(bid_at_trade - ask_at_trade),5)
print(f'Bid at Trade: {bid_at_trade}')
print(f'Ask at Trade: {ask_at_trade}')
print(f'Spread at Trade: {spread_at_trade}')

# Get historical prices
start_time = datetime.now() - relativedelta(days=30)
start_time = time.mktime(start_time.timetuple())
end_time = time.mktime(datetime.now().timetuple())

header = {'Authorization': 'Bearer ' + token}
query = {'from': str(start_time), 'to': str(end_time), 'granularity': interval}
candles_path = f'/v3/accounts/{acct}/instruments/{currency}/candles'
response = requests.get('https://' + rest_url + candles_path,
                        headers=header,
                        params=query)
print(f'Response to historical query: {response}')

df = json_to_pandas(response)
```
